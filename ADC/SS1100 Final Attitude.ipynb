```python
def calculate_rotation(current_orientation, desired_orientation): #searched on stack overflow for rotation monitoring, found one about a sphere and used that as my base
    """Calculate the rotation required to achieve desired orientation"""
    rotation = tuple(desired - current for current, desired in zip(current_orientation, desired_orientation))
    return rotation
# Test #Open AI 4.0 used to determine how to format since I could not get the print statement at the bottom to work 
current_orientation = (30, 60, 90)
desired_orientation = (0, 10, 15)
rotation_needed = calculate_rotation(current_orientation, desired_orientation)
print(f"Rotation needed: {rotation_needed}")

# save rotation to a file
file_path = "rotation_output.txt"
with open(file_path, "w") as file:
    file.write(f"Rotation_needed: {rotation_needed}")
print(f"Rotation needed: {rotation_needed} (saved to {file_path})")
import math #asked OPEN AI "How do I add math and random numbers"
import random

class AttitudeControlSystem:
    """
    Represents the Attitude Control Subsystem (ACS).
    """
    def __init__(self): #watched a youtube video on __init__ and then asked OpenAI to show me an example of a tupple using basic numbers. Swapped out numbers for the object in this code
        self.current_orientation = (0, 0, 0)  # Default initial orientation (X, Y, Z)

    def calculate_rotation(self, current_orientation, desired_orientation): #Asked how "define the orientation of a sattalite in a code as a tupple"
        """
        Calculate the rotation required to achieve the desired orientation.

        Parameters:
            current_orientation (tuple): Current orientation (X, Y, Z). #this is from Open Ai
            desired_orientation (tuple): Desired orientation (X, Y, Z).

        Returns:
            tuple: Rotation needed (delta_X, delta_Y, delta_Z).
        """
        return tuple(desired - current for current, desired in zip(current_orientation, desired_orientation)) #I asked OPEN AI to show me how ow to save a file

    def apply_rotation_with_error(self, rotation):
        """
        Apply the calculated rotation to update the current orientation with random error.

        Parameters:
            rotation (tuple): Rotation to apply (delta_X, delta_Y, delta_Z).
        """
        error = tuple(random.uniform(-0.05, 0.05) for _ in rotation)
        self.current_orientation = tuple(current + delta + err for current, delta, err in zip(self.current_orientation, rotation, error))

    def is_within_tolerance(self, desired_orientation, tolerance=0.1):
        """
        Check if the current orientation is within the desired tolerance.

        Parameters:
            desired_orientation (tuple): Desired orientation (X, Y, Z).
            tolerance (float): Allowed tolerance for each axis.

        Returns:
            bool: True if within tolerance, False otherwise.
        """
        return all(abs(current - desired) <= tolerance for current, desired in zip(self.current_orientation, desired_orientation)) #I pasted this into Open AI and asked to trouble shoot. I was doing it as a print statement at first 

    def read_current_state_from_file(self, file_path):
        """
        Read the current orientation from a file.

        Parameters:
            file_path (str): Path to the file containing the current state.
        """
        with open(file_path, 'r') as file:
            self.current_orientation = tuple(map(float, file.read().strip().split(',')))

    def write_current_state_to_file(self, file_path):
        """
        Write the current orientation to a file.

        Parameters:
            file_path (str): Path to the file to write the current state.
        """
        with open(file_path, 'w') as file:
            file.write(','.join(map(str, self.current_orientation)))


# Main program to test the class I asked OPEN AI to show me how to test since I was failing with print statements
def main():
    acs = AttitudeControlSystem()
    state_file_path = "current_state.txt"

    # Initialize the state file if it doesn't exist
    try:
        acs.read_current_state_from_file(state_file_path)
    except FileNotFoundError:
        with open(state_file_path, 'w') as file:
            file.write("0,0,0")  # Default initial orientation
        acs.read_current_state_from_file(state_file_path)

    # Target orientations
    target_orientations = [
        (100, 200, 300),
        (0, 0, 0),
        (3, 30, 300)
    ]

    # Iterate through each target orientation
    for target in target_orientations:
        print(f"\nAdjusting to target orientation: {target}")
        while not acs.is_within_tolerance(target):
            rotation_needed = acs.calculate_rotation(acs.current_orientation, target)
            print(f"Current orientation: {acs.current_orientation}, Rotation needed: {rotation_needed}")
            acs.apply_rotation_with_error(rotation_needed)
            acs.write_current_state_to_file(state_file_path)

        print(f"Target orientation reached: {acs.current_orientation}")


if __name__ == "__main__":
    main()
```
